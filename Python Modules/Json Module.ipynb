```python
import json5
```

```python
people_string = '''
{
    "people":[
        {
            "name":"Nikhil Sawant",
            "phone_no": "2635723676",
            "emails": ["nikhilsawant6576@xmail.com","killfr46574@vmail.com"],
            "has_license": false
            
        },
        {
            "name":"Jay Sawant",
            "phone_no": "266784676",
            "emails": null,
            "has_license": true
        }
    ]

}
'''
```

```python
data = json5.loads(people_string)

print(data)
```

```python
print(type(data))
```

```python
data['people']
```

```python
data['people'][0]['name']
```

```python
data['people'][1]['has_license']
```

```python
for person in data['people']:
    print(person)
```

```python
for person in data['people']:
    del person['phone_no']
```

```python
data
```

```python
new_string = json5.dumps(data, indent=2, sort_keys=True)

print(new_string)
```

### Working with json file now

```python
with open("valorant.json") as f:
    data = json5.load(f)
```

```python
data
```

```python
for char in data['characters']:
    print(char['name'], char['id'])
```

```python
data
```

```python
for char in data['characters']:
    del char['localizedNames']
```

```python
with open('valorant_tweets_cleaned.json', 'w') as v:
    json5.dump(data, v, indent=2)
```
